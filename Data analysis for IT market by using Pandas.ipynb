### import requaird library 

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import os 
```

### check the current diractory and read all files that inside it

```python
cwd = os.path.abspath('') 
files = os.listdir(cwd) 
cwd
files
```

### merge all the files for diffrent months in one CSV file

```python
df = pd.DataFrame()
for file in files:
     if file.endswith('.csv'):
         df = df.append(pd.read_csv(file), ignore_index=True) 
df.head()
df.to_csv("all_data.csv", index=False)
```

```python
df.head()
```

### Cleaning data 

```python
df_NAN = df[df.isnull().any(axis=1)]
df_NAN

```

```python
df = df.dropna(how='all')
df.head()
```

### clean the data from OR 

```python
new_df = df[df['Order Date'].str[0:2] != 'Or']
new_df.head()
```

### Change the quantity and price column from str to int

```python
new_df['Quantity Ordered']= pd.to_numeric(new_df['Quantity Ordered'])
new_df['Price Each']= pd.to_numeric(new_df['Price Each'])
new_df.head()
```

### Add column of months to the table

```python
import warnings 
warnings.filterwarnings("ignore")
new_df['Months'] = new_df['Order Date'].str[0:2]
new_df['Months'] = new_df['Months'].astype('int32')
df.head()
```

### What is the best months of sales ?

```python
new_df['Sales'] = new_df['Quantity Ordered']*new_df['Price Each']
new_df.head()
```

```python
answer_1= new_df.groupby('Months').sum()
answer_1
```

```python
months = range(1,13)
plt.bar(months,answer_1['Sales'])
plt.xlabel('Minths')
plt.ylabel('Sales in USD$')
plt.show()
```

### What city had the heighest sales for the prodacts?

#### Add a city column to the current table¶

```python
answer_2=new_df['cities']= new_df['Purchase Address'].apply(lambda x: x.split(',')[1])
new_df.head()
```

### What is the best time for making adv for products 

```python
new_df.head()
```

```python
new_df['Order Date']=pd.to_datetime(new_df['Order Date'])##CHange the type of Order Date to Datatime
new_df.head()
```

```python
new_df['Hour']=new_df['Order Date'].dt.hour
new_df.head()
```

```python
Hours = [Hour for Hour, df in new_df.groupby("Hour")]
plt.plot(Hours, new_df.groupby(['Hour']).count())
plt.xticks(Hours)
plt.xlabel('Hours')
plt.ylabel('Number of order')
plt.show()
```
